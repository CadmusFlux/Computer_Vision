```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Steps:

1) Read image

2) Create a matrix of the same shape as the image

3) Create a noise matrix with  [Mean = 0] and [S.D = 20]

4) Overalp  Image matrix and noise matrix to get a smoothened image.

5) Fin :))

```
image = cv2.imread("a.jpg")

dst = np.empty_like(image)

noise = cv2.randn(dst,(0,0,0),(20,20,20))
img_noise = cv2.addWeighted(image,0.7,noise,0.3,0)
```

```
plt.imshow(img_noise)
plt.show()
```
